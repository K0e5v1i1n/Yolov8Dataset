```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_path = '/content/gdrive/MyDrive/Yolov8Developer/Project_Yolov8/runs/detect/train2/results.csv'
```

```python
results = pd.read_csv(results_path)
```

```python
print(results.keys())
```

```python
plt.figure()
plt.plot(results['                  epoch'], results['         train/box_loss'], label='train loss')
plt.plot(results['                  epoch'], results['           val/box_loss'], label='val loss', c='red')
plt.grid()
plt.title('Loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend()
plt.savefig('/content/gdrive/MyDrive/Yolov8Developer/Project_Yolov8/analyze/Loss.png')
```

```python
plt.figure()
plt.plot(results['                  epoch'], results['   metrics/precision(B)'] * 100)
plt.grid()
plt.title('Validation precision')
plt.ylabel('precision (%)')
plt.xlabel('epochs')
plt.savefig('/content/gdrive/MyDrive/Yolov8Developer/Project_Yolov8/analyze/precision.png')
```

```python
plt.figure()
plt.plot(results['                  epoch'], results['      metrics/recall(B)'] * 100)
plt.grid()
plt.title('Validation recall')
plt.ylabel('recall (%)')
plt.xlabel('epochs')
plt.savefig('/content/gdrive/MyDrive/Yolov8Developer/Project_Yolov8/analyze/recall.png')
```
